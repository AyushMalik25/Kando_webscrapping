```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Initialize an empty list to store all data
all_data = []

# Base URL of the website (without page number)
base_url = "https://kando.tech/investors/all?page=0"  # Replace with the actual base URL

# Loop through all 650 pages
for page in range(1, 633):  # Adjust the range to 651 for 650 pages
    url = base_url + str(page)  # Construct the full URL
    print(f"Scraping data from page {page}: {url}")

    # Fetch the webpage
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract data from the current page
        data = []
        table = soup.find('table')  # Adjust selector for the table
        if table:  # Check if the table exists
            rows = table.find_all('tr')  # Find all rows

            for row in rows:
                cols = row.find_all('td')  # Find all columns
                cols = [col.text.strip() for col in cols]  # Extract text
                data.append(cols)

            # Append the data from the current page to the main list
            all_data.extend(data)
            print(f"Successfully scraped {len(data)} rows from page {page}")
        else:
            print(f"No table found on page {page}")
    else:
        print(f"Failed to retrieve page {page}. Status code: {response.status_code}")

    # Add a delay to avoid overloading the server
    time.sleep(1)  # Adjust the delay as needed

# Convert the data to a DataFrame
df = pd.DataFrame(all_data)

# Save the data to an Excel file
output_file = "scraped_data_633_pages.xlsx"
df.to_excel(output_file, index=False, header=False)
print(f"All data saved to {output_file}")
```
